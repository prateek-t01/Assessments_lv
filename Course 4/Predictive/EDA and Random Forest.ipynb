```python
def detect_and_treat_outliers(df, columns):
    """
    Detects and treats outliers in the specified columns of a dataframe.
    Outliers are replaced with the median of the column.

    Parameters:
    df (pandas.DataFrame): The dataframe containing the data.
    columns (list): List of column names to check for outliers.

    Returns:
    pandas.DataFrame: The dataframe with outliers treated.
    """

    for col in columns:
        # Calculate Q1, Q3 and IQR
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        # Define bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Replace outliers with the median of the column
        median = df[col].median()
        df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound), median, df[col])

    return df

columns = ['carat', 'depth', 'table', 'x', 'y', 'z', 'price']
df = detect_and_treat_outliers(df, columns)
```

```python
# Model Traning
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# spilit Train and Test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    r2 = r2_score(y_true, y_pred)
    return mae, rmse, r2
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest Regressor': RandomForestRegressor(),
    'K-Nearest Neighbors Regressor': KNeighborsRegressor()
}

model_list = []
r2_list = []

for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mae, rmse, r2_square = evaluate_model(y_test, y_pred)

    print(model_name)
    model_list.append(model_name)
    print('Model Performance')
    print("RMSE:", rmse)
    print("MAE:", mae)
    print("R2 Score:", r2_square * 100)
    r2_list.append(r2_square)
    print('=' * 35)

print('')
print("Best Performing Model:", model_list[r2_list.index(max(r2_list))], '\nR2 Score:', max(r2_list) * 100)
```
